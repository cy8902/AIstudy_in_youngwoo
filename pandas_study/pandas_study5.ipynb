```python
# %load fonts.py
#!/usr/bin/env python

# In[ ]:


# matplotlib 한글 폰트 문제 해결
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import rc
import platform 

if platform.system() == 'Darwin':   # Mac 환경 폰트 지정
    plt.rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    rc('font', family='Malgun Gothic')
    
plt.rc('axes', unicode_minus=False)

```

```python
import seaborn as sns
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
titanic.info()
```

```python
fig = plt.figure(figsize = (15,5))

ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

sns.regplot(x = "age",y = "fare", data = titanic, ax = ax1,color = "red",
           marker = ".")
sns.regplot(x = "age",y = "fare", data = titanic, ax = ax2,
           fit_reg =False)


ax1.set_title("선형회귀선이 있는 그래프")
ax1.set_xlabel("나이")
ax1.set_ylabel("요금")
plt.show()
```

```python
fig = plt.figure(figsize = (15,5))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

sns.distplot(titanic["fare"], ax = ax1)
sns.kdeplot(x = "fare",data = titanic, ax = ax2)
sns.histplot(x = "fare",data = titanic, ax = ax3)

ax1.set_title("titanic fare - hist /ked")
ax2.set_title("titanic fare - ked")
ax3.set_title("titanic fare - hist ")

plt.show()
```

```python
# 클래스별로 성별을 나눠
table = titanic.pivot_table(index = ["sex"], columns = ["class"],
                           aggfunc = "size")


sns.heatmap(table,)
```

```python
titanic["class"].unique()
titanic.sex.unique()
```

```python
fig = plt.figure(figsize = (25,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.stripplot(x = "class",y = "age",data = titanic,ax = ax1)
sns.swarmplot(x = "class",y = "age",data = titanic,ax = ax2)

ax1.set_title("Strip plot")
ax2.set_title("Swarm plot")
plt.show()
```

```python
# 막대 그래프
fig = plt.figure(figsize = (15,5))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

sns.barplot(x = "sex",y = "survived", data = titanic, ax = ax1)
sns.barplot(x = "sex",y = "survived", data = titanic, hue ="class", ax = ax2)
sns.barplot(x = "sex",y = "survived", data = titanic, hue ="class", 
            dodge = False, ax = ax3)

```
