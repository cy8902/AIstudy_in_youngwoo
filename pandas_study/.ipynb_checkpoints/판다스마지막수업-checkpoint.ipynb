```python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset("titanic")
df = titanic.loc[:,["age","fare"]]

df["ten"] = 10
df
```

```python
def add_10(n):
    return n+10
def add_two_obj(a,b):
    return a + b
```

```python
sr1 = df["age"].apply(add_10)
sr1
sr2 = df["age"].apply(add_two_obj,b = 10)
sr2.head()
sr3 = df["age"].apply(lambda x : add_10(x))
sr3

df_map = df.applymap(add_10)
df_map.head()
```

```python
def missing_value(series):
    return series.isnull()

result = df.apply(missing_value, axis = 0)
result
```

```python
def min_max(x):
    return x.max() - x.min()

df_1 = df.apply(min_max)  # 시리

# print(df.apply(min_max))
df_1
```

```python
df = titanic.loc[:,["age","sex","class","fare",
                 "survived"]]
df
```

```python
print("승객 수:",len(df))
print(df.head())

print()

grouped = df.groupby(["class"])
print(grouped)

for key, group in grouped:
    print("key :",key)
    print("number :",len(group))
    print(group.head())
    print("\n")
```

```python
avg = grouped.mean()
avg
```

```python
g3 = grouped.get_group("Third")
g3
```

```python
group_two = df.groupby(["class","sex"])

for key, group in group_two:
    print("key :",key)
    print("number :",len(group))
    print(group.head())
```

```python
av2 = group_two.mean()
print(av2)
print("")
```

```python
agg_group = grouped.agg(["min","max"])
agg_group

print("")
agg_group2 = grouped.agg({"fare": ["min","max"],
                          "age": "mean"})
agg_group2
```

```python
def z_score(x):
    return (x - x.mean())/x.std()

age_zscore = grouped.age.transform(z_score)

print(age_zscore.loc[[1, 9, 0]])
print("")
print(len(age_zscore))
print("")
print()
print("")
```
